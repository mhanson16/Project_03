```python
def get_year_df(year):
    df_year = df.loc[df['Year']==year].reset_index(drop=True)
    return df_year
```

```python
df_2021 = get_year_df(2021)
df_2021.head()
```

```python
df_2021_sorted=df_2021.sort_values(by=['Total'], ascending=False)
df_2021_sorted.head(10)
```

```python
df_2020 = get_year_df(2020)
df_2020.head()
```
