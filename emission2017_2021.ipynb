```python
## new json for pie
piedata = pd.read_csv('pie.csv')
ID = ['coal', 'oil', 'gas', 'cement', 'flaring', 'other']
year = ["2017", "2018", "2019", "2020", "2021"]
years = [2017, 2018, 2019, 2020, 2021]
piedata.loc[piedata['iso']=='USA'].reset_index(drop=True)
```

```python
# coal
df = piedata
df = df[['country', 'year', 'coal']]

dict_coal = {}
dict_coal_data = {}

for i in range(len(year)):
    # get the data
    data = df.loc[df['year']==years[i]].reset_index(drop=True)
    dict = {}
    for x in range(len(data.country)):
        dict[data.country[x]] = data.coal[x]
    dict_coal_data[year[i]] = dict
    
dict_coal['id'] = "coal"
dict_coal['metadata'] = dict_coal_data
dict_coal
```
