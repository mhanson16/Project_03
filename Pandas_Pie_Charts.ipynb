```python
import pandas as pd 
import matplotlib as plt
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import numpy as np
from matplotlib import pyplot as plt

```

```python
# Reading the csv file
df_new = pd.read_csv('Initial_Data/GCB2022v27_percapita_flat.csv')
df_new
```

```python
# saving xlsx file
GFG = pd.ExcelWriter('PF_Files/GCB2022v27_percapita_flat.xlsx')
df_new.to_excel(GFG, index=False)
GFG.save()
```

```python
#Biggest Contributiors of CO2 Emission by Coal

Coal_df_new = df_new[["Country","Coal", "Year","Total"]] 
top_Coal_df_new= Coal_df_new.nlargest(5, 'Coal')
top_Coal_df_new
Coal_title = 'CO2 Emissions caused by Coal'
top_Coal_df_new.groupby(['Year','Country', 'Total']).sum().plot(kind='pie',y='Coal',fontsize=6, shadow=True,label="", figsize=(2,2),title = Coal_title, autopct='%1.0f%%', legend = None,explode = (0, 0, 0.10, 0, 0) )
```
