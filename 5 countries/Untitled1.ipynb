```python
import pandas as pd
import json
```

```python
df = pd.read_csv('five_countries.csv')
df.head()
```

```python
# # just check the data in 2021
df2021 = df.loc[df['year']==2021].reset_index(drop=True)
df2021
```

```python
dict = {}

# convet the following columns to dictonary
stat = ['coal', 'oil', 'gas', 'cement', 'flaring', 'other']
stat_dict = df2021[stat].to_dict('records')

for i in range(len(df2021.country)):
    dict[df2021.country[i]] = stat_dict[i]
dict
```

```python
# Serializing json  
json_object = json.dumps(dict, indent = 4) 

# Save to json file
with open("five_countries2021.json", "w") as test_file:
    test_file.write(json_object)
```
